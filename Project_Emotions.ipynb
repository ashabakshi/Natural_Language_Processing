```python
#  Convert emotion labels (text) into numbers (Label Encoding)
# ML models cannot understand text, so we convert categories → numerical values

# Step 1: Get unique emotion labels
unique_emotions=df['Emotions'].unique()


# Step 2: Create mapping dictionary (emotion → number)
emotion_numbers={}
i=0
for emo in unique_emotions:
    emotion_numbers[emo]=i
    i+=1
# Example mapping:
# {'happy': 0, 'sad': 1, 'angry': 2}

# Step 3: Replace text labels with numbers in dataframe
df['Emotions']=df['Emotions'].map(emotion_numbers)
```
