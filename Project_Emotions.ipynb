```python
#  Convert emotion labels (text) into numbers (Label Encoding)
# ML models cannot understand text, so we convert categories → numerical values

# Step 1: Get unique emotion labels
unique_emotions=df['Emotions'].unique()


# Step 2: Create mapping dictionary (emotion → number)
emotion_numbers={}
i=0
for emo in unique_emotions:
    emotion_numbers[emo]=i
    i+=1
# Example mapping:
# {'happy': 0, 'sad': 1, 'angry': 2}

# Step 3: Replace text labels with numbers in dataframe
df['Emotions']=df['Emotions'].map(emotion_numbers)
```

```python
# lets clean Text now
# lower case 
df['text']= df['text'].apply(lambda x: x.lower())
```

```python

# removing punctuation

import string 

def remove_punc(txt):
    return txt.translate(str.maketrans('','',string.punctuation))
# apply on text column
df['text']=df['text'].apply(remove_punc) 

    
```

```python
# removing numbers 
def remove_numb(txt):
    new =""
    for i in txt:
        if not i.isdigit():
            new= new + i 
    return new 
# apply on text col for removing num 
df['text']=df['text'].apply(remove_numb)
```

```python
# removing emojis 

def remove_emojis(txt):
    new =""
    for i in txt:
        if i.isascii():
            new= new + i 
    return new 
# removing emoji 
# apply on text col for removing num 

df['text']=df['text'].apply(remove_emojis)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# stopwords
stop_words = set(stopwords.words('english'))
# stop_words
```

```python
df.loc[1]['text']
```

```python
# now lets do tokenization for removing stop words 

def remove_stopwords(txt):
  words = txt.split() #tokenization
  
  cleaned = []
  for i in words:
    if not i in stop_words: # remove stop words 
      cleaned.append(i)

  return ' '.join(cleaned) # join back to sentence

  # apply
df['text'] = df['text'].apply(remove_stopwords)
```
